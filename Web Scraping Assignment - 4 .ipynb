Question-1: Wikipedia.

```python
# Importing libraries.
import selenium
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd
```

```python
# Driver
driver = webdriver.Chrome(r"C:\Users\HOMA\Downloads\chromedriver_win32\chromedriver.exe")
```

```python
# Website url.
url='https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'
```

```python
# Opening website url
driver.get(url)
```

```python
# Creating Empty sets.
Rank=[]
Title=[]
Artist=[]
View=[]
Date=[]
```

```python
# Scraping the data.
rank=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[1]')
title=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[2]')
artist=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[3]')
view=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[4]')
date=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[5]')
```

```python
#iterate over the elements to get the required data
for i in [rank,title,artist,view,date]:
    for j in i:
        if i == rank:
            Rank.append(j.text)
        if i == title:
            Title.append(j.text)
        if i== artist:
            Artist.append(j.text)
        if i == view:
            View.append(j.text)
        if i == date:
            Date.append(j.text)
```

```python
# Creating DataFrame.
youtube_Wiki=pd.DataFrame({})
youtube_Wiki['Rank']=Rank
youtube_Wiki['Title']=Title
youtube_Wiki['Artist']=Artist
youtube_Wiki['Views']=View
youtube_Wiki['Date']=Date
```

```python
# Creating Empty sets.
Data_Name = []
Data_type = []
Task = []
Attribute_type = []
Instance = []
Attribute = []
Year = []
```

```python
#scraping the data for all names.
name = driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td/table/tbody/tr/td[2]')
data_type = driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[2]/p')
task = driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[3]/p')
attribute_type = driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[4]/p')
instance = driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[5]/p')
attribute = driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[6]/p')
year = driver.find_elements_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[7]/p')
```

```python
# Appending the data into text through for loop.
for i in name:
    Data_Name.append(i.text)
for i in [data_type,task,attribute_type,instance,attribute,year]:
    for j in i[1:]:
        if i==data_type:
            Data_type.append(j.text)
        if i==task:
            Task.append(j.text)
        if i==attribute_type:
            Attribute_type.append(j.text)
        if i==instance:
            Instance.append(j.text)
        if i==attribute:
            Attribute.append(j.text)
        if i==year:
            Year.append(j.text)
```

Question-5: GitHub

```python
# URL of the website.
url = 'https://github.com/'
```

```python
# Getting URL.
driver.get(url)
```

```python
# Clicking on the trending option through Explore bar.
Trending = driver.find_element_by_xpath("//ul[@class='list-style-none mb-3']/li[3]/a")
driver.get(Trending.get_attribute('href'))
```

```python
# Creating empty lists.
Repository_title=[]
Repository_description=[] 
Contributors_count=[] 
Language_used=[]
```

```python
# Finding elements and appending through NoSuchElementException.
repository_title = driver.find_elements_by_xpath('//h1[@class="h3 lh-condensed"]')
for i in repository_title:
    Repository_title.append(i.text)

description = driver.find_elements_by_xpath('//p[@class="col-9 text-gray my-1 pr-4"]')
for i in description:
    Repository_description.append(i.text)

language_used = driver.find_elements_by_xpath('//span[@class="d-inline-block ml-0 mr-3"]/span[2]')
for i in language_used:
    
    try:
        Language_used.append(i.text)
        
    except NoSuchElementException :
        
        Language_used.append("---")
        
urls = []
for i in driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']/a"):
    urls.append(i.get_attribute("href"))

for URL in urls:
    
    try:
        driver.get(URL)
        count = driver.find_element_by_xpath("//div[@class='BorderGrid-cell']/h2/a/span").text
        Contributors_count.append(count)
    except NoSuchElementException :
        Contributors_count.append("---")
```

```python
# Creating DataFrame.
GitHub_Repositories = pd.DataFrame({})
GitHub_Repositories['Repository_Title'] = Repository_title[:23]
GitHub_Repositories['Repository_Description'] = Repository_description[:23]
GitHub_Repositories['Contributors_Count'] = Contributors_count[:23]
GitHub_Repositories['Language_Used'] = Language_used[:23]
```
